```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
f = open('demo.html','r',encoding='utf-8')
res = f.read()
f.close()
```

```python
soup = BeautifulSoup(res,'html.parser')
```

```python
d = {"Name":[],"ROM":[],"RAM":[],"color":[],"star":[],"review":[],"real camera":[],"front_camera":[],"old_price":[],"New_price":[],
     "discount":[],"rating":[],"display(cm)":[],"display(inch)":[]}
mobiles = soup.find_all('div',class_="yKfJKb row")
print(len(mobiles))
for i in mobiles:
    d["Name"].append(i.find('div',class_="KzDlHZ").text.strip().split('(')[0].strip())
    d["color"].append(i.find('div',class_="KzDlHZ").text.strip().split('(')[1].split(',')[0])
    d["ROM"].append(i.find('div',class_="KzDlHZ").text.strip().split(',')[1].strip().split(')')[0])
    d["star"].append(i.find('div',class_="XQDdHH").text.strip())
    d["rating"].append(i.find('span',class_="Wphh3N").find('span').find_all('span')[0].text.split()[0])
    d["review"].append(i.find('span',class_="Wphh3N").find('span').find_all('span')[2].text.split()[0])
    # d["RAM"].append(i.find('ul',class_="G4BRas").find_all('li',class_='J+igdf')[0].text.split('|')[0])
    d["RAM"].append(i.find('ul',class_="G4BRas").text.split('|')[0].strip())
    d['display(cm)'].append(i.find('ul',class_="G4BRas").text.split("|")[1].split('ROM')[1].split('(')[0])
    d['display(inch)'].append(i.find('ul',class_="G4BRas").text.split("|")[1].split("ROM")[1].split("(")[1].split('inch)')[0])
    d['real camera'].append(i.find('ul',class_="G4BRas").text.split("|")[1].split("Display")[1])
    d['front_camera'].append(i.find('ul',class_="G4BRas").text.split("|")[2].split()[0])
    d['old_price'].append(i.find('div',class_="Nx9bqj _4b5DiR").text[1:].replace(",",""))
    try:
        d['discount'].append(i.find('div',class_="UkUFwK").text.split()[0])
        d["New_price"].append(i.find('div',class_="yRaY8j ZYYwLA").text[1:].replace(",",""))
    except:
        d['discount'].append('NaN')
        d["New_price"].append('NaN')
df = pd.DataFrame(d)
df
```
